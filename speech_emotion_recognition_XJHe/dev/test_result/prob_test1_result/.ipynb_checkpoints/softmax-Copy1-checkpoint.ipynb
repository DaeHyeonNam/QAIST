```python
import numpy as np
import csv
    
def softmax(x):
    
    """Compute softmax values for each sets of scores in x."""
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
with open("./result2.csv" , mode= 'r') as csvfile:
    prob = csv.reader(csvfile, delimiter=',')
    prob = list(prob)
    for i in range(len(prob)):
        for j in range(len(prob[0])):
            prob[i][]
    
    for i in range(len(prob)):
        temp =softmax(prob[i])
        with open("./processed.csv", 'a') as writeFile:
            writer = csv.writer(writeFile)
            writer.writerow(temp)
```
