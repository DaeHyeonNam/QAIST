```python
import numpy as np
import tensorflow as tf
from acrnn1 import acrnn
from extract_mel import next_batch, test_dataset
import cPickle
import csv

from sklearn.metrics import precision_score as precision
from sklearn.metrics import recall_score as recall
from sklearn.metrics import confusion_matrix as confusion
import os
```

```python
# testWavFolder = "/mnt/home/20150132/Downloads/wavData/"
testWavFolder = "/mnt/home/20150132/Desktop/dataset/wav/val/"
outputfile = "./result.csv"
ckptfile = './checkpoint/checkpoint_v2/model4.ckpt-7166'

fromLabel = {0:'hap', 1:'sur', 2:'dis', 3:'neu', 4:'fea', 5:'sad', 6:'ang'}

def getFilesList(folder):
    return [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]

testWavFiles = getFilesList(testWavFolder)
numTestFiles = len(testWavFiles)

testWavFiles.sort()
```

```python
tf.app.flags.DEFINE_integer('num_epoch', 300000, 'The number of epoches for training.')
tf.app.flags.DEFINE_integer('num_classes', 7, 'The number of emotion classes.')
tf.app.flags.DEFINE_integer('batch_size', 45, 'The number of samples in each batch.')
tf.app.flags.DEFINE_boolean('is_adam', True, 'whether to use adam optimizer.')
tf.app.flags.DEFINE_float('learning_rate', 0.00001, 'learning rate of Adam optimizer')
tf.app.flags.DEFINE_float('dropout_keep_prob', 1, 'the prob of every unit keep in dropout layer')
tf.app.flags.DEFINE_integer('image_height', 300, 'image height')
tf.app.flags.DEFINE_integer('image_width', 40, 'image width')
tf.app.flags.DEFINE_integer('image_channel', 3, 'image channels as input')
tf.app.flags.DEFINE_integer('val_iter', 5, 'the number of validation test')

tf.app.flags.DEFINE_string('checkpoint', './checkpoint/', 'the checkpoint dir')
tf.app.flags.DEFINE_string('model_name', 'model4.ckpt', 'model name')
tf.app.flags.DEFINE_string('f', '', 'kernel')

FLAGS = tf.app.flags.FLAGS

test_batch = 45
```

```python
def test():
    with open(outputfile, 'a') as writeFile:
        writer = csv.writer(writeFile)
        writer.writerow(["FileID", "Emotion"])

    X = tf.placeholder(tf.float32, shape=[None, FLAGS.image_height,FLAGS.image_width,FLAGS.image_channel])
    
    is_training = tf.placeholder(tf.bool)

    keep_prob = tf.placeholder(tf.float32) # drop out rate
    
    Ylogits = acrnn(X, is_training=is_training, dropout_keep_prob=keep_prob) # result
    
    with tf.Session() as sess:
        saver = tf.train.Saver()
        saver = tf.train.import_meta_graph(ckptfile+'.meta')
        saver.restore(sess, ckptfile)
        
        i =0
        while(i < numTestFiles):
            if(i+test_batch >numTestFiles):
                test_data, pernums = test_dataset(testWavFolder,testWavFiles, i, numTestFiles- i)
            else:
                test_data, pernums = test_dataset(testWavFolder,testWavFiles, i, test_batch)
            i+=test_batch
            
            print(pernums.shape)
            y_pred_valid = sess.run(Ylogits,feed_dict = {X:test_data, is_training:False, keep_prob:1})

            
            index =0
            for s in range(pernums.shape[0]):
                y_test = np.max(y_pred_valid[index: index+ pernums[s]],0)
                
                test_result = np.argmax(y_test)
                print(test_result)
                
                with open(outputfile, 'a') as writeFile:
                    writer = csv.writer(writeFile)
                    writer.writerow([fromLabel[int(test_result)] ])

                index = index + pernums[s]
                
    
```

```python
test()
```
